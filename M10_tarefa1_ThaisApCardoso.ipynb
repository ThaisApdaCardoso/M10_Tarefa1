Uma instituição financeira quer conhecer melhor o perfil de renda de seus novos clientes para diversos fins, por exemplo, melhor dimensionar o limite de cartões de crédito dos novos clientes, sem necessariamente solicitar olerites ou documentações que impactem na experiência do seu cliente.

Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
````

````python
arquivo = 'previsao_de_renda.csv'
dados = pd.read_csv(arquivo)
````

````python
# Matriz de dispersão
sns.pairplot(dados)
plt.show()
````

````python
# Matriz de correlação
dados_numericos = dados.select_dtypes(include=['float64', 'int64'])
correlacao = dados_numericos.corr()

# Visualizando a matriz de correlação com mapa de calor
plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
````

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
# Criando o gráfico de dispersão para as duas variáveis mais correlacionadas com renda
plt.figure(figsize=(10,6))

# Gráfico de dispersão para a primeira variável (tempo_emprego)
sns.scatterplot(x=dados['tempo_emprego'], y=dados['renda'], label='tempo_emprego')

# Gráfico de dispersão para a segunda variável (idade)
sns.scatterplot(x=dados['idade'], y=dados['renda'], label='idade')

# Adicionando rótulos e título ao gráfico
plt.xlabel('Variáveis Mais Correlacionadas')
plt.ylabel('Renda')
plt.title('Gráfico de Dispersão - Variáveis mais correlacionadas com Renda')
plt.legend()
plt.show()
````

3) Na su aopinião, há outliers na variável renda?

````python
# Verificando se há outliers na variável renda com um boxplot
plt.figure(figsize=(8,6))
sns.boxplot(y=dados['renda'])
plt.title('Boxplot da Renda - Identificação de Outliers')
plt.ylabel('Renda')
plt.show()
````

No boxplot da variável renda acima, podemos observar que existem alguns pontos isolados acima da caixa, o que indica a presença de outliers (valores que estão bem acima do esperado).

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
# Calculando o logaritmo da variável renda
dados['log_renda'] = np.log(dados['renda'])

# Repetindo a análise com o gráfico de dispersão para as variáveis mais correlacionadas
plt.figure(figsize=(10,6))

# Gráfico de dispersão para a primeira variável (tempo_emprego) com a renda logaritimizada
sns.scatterplot(x=dados['tempo_emprego'], y=dados['log_renda'], label='tempo_emprego')

# Gráfico de dispersão para a segunda variável (idade) com a renda logaritimizada
sns.scatterplot(x=dados['idade'], y=dados['log_renda'], label='idade')

# Adicionando rótulos e título ao gráfico
plt.xlabel('Variáveis Mais Correlacionadas')
plt.ylabel('Logaritmo da Renda')
plt.title('Gráfico de Dispersão - Logaritmo da Renda')
plt.legend()
plt.show()
````

````python
# Repetindo o boxplot com a renda logaritimizada
plt.figure(figsize=(8,6))
sns.boxplot(y=dados['log_renda'])
plt.title('Boxplot da Logaritmo da Renda - Identificação de Outliers')
plt.ylabel('Logaritmo da Renda')
plt.show()
````

5) A aplicação do LOG você acha que melhorou ou piorou a análise?

Pode-se notar que a aplicação do logaritmo melhorou a análise, porque ajudou a reduzir o impacto dos outliers e tornou os dados mais fáceis de trabalhar visualmente e para a modelagem.
